```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

food = ctrl.Antecedent(np.arange(0, 11, 1), 'Food')
Service = ctrl.Antecedent(np.arange(0, 11, 1), 'Service')
tip = ctrl.Consequent(np.arange(10, 31, 1), 'tip')

# tao cac tam giac 

food['poor'] = fuzz.trimf(food.universe, [0, 0, 5])
food['average'] = fuzz.trimf(food.universe, [0, 5, 10])
food['good'] = fuzz.trimf(food.universe, [5, 10, 10])

Service['poor'] = fuzz.trimf(Service.universe, [0, 0, 5])
Service['average'] = fuzz.trimf(Service.universe, [0, 5, 10])
Service['good'] = fuzz.trimf(Service.universe, [5, 10, 10])

tip['less'] = fuzz.trimf(tip.universe, [10, 10, 20])
tip['normal'] = fuzz.trimf(tip.universe, [10, 20, 30])
tip['much'] = fuzz.trimf(tip.universe, [20, 30, 30])

#----------------------------------------------------------------------------------
#-------------rule-------------------------

rule  = ctrl.Rule(food['poor'] & Service['poor'], tip['less'])
rule1 = ctrl.Rule(food['poor'] & Service['average'], tip['normal'])
rule2 = ctrl.Rule(food['poor'] & Service['good'], tip['normal'])
rule3 = ctrl.Rule(food['average'] & Service['poor'], tip['normal'])
rule4 = ctrl.Rule(food['average'] & Service['average'], tip['normal'])
rule5 = ctrl.Rule(food['average'] & Service['good'], tip['normal'])
rule6 = ctrl.Rule(food['good'] & Service['poor'], tip['normal'])
rule7 = ctrl.Rule(food['good'] & Service['average'], tip['normal'])
rule8 = ctrl.Rule(food['good'] & Service['good'], tip['much'])
#------------------------------------------------------------------------------------
#---------------training--------------------------------------------------

tipping_ctrl = ctrl.ControlSystem([rule, rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8])
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)
tipping.input['Food'] = 6.5
tipping.input['Service'] = 9.8
tipping.compute()
print(tipping.output['tip'])
tip.view(sim = tipping)


```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

Rice = ctrl.Antecedent(np.arange(0, 1000, 50), 'Rice')
Time = ctrl.Antecedent(np.arange(5, 180, 10), 'Time')
power = ctrl.Consequent(np.arange(10, 100, 1), 'power')

# tao cac tam giac 

Time['veryfast'] = fuzz.trimf(Time.universe, [10, 15, 20])
Time['fast'] = fuzz.trimf(Time.universe, [20, 30, 40])
Time['avg'] = fuzz.trimf(Time.universe, [40, 60, 80])
Time['slow'] = fuzz.trimf(Time.universe, [80, 100, 120])
Time['veryslow'] = fuzz.trimf(Time.universe, [120, 150, 180])

Rice['verylow'] = fuzz.trimf(Rice.universe, [200, 250, 300])
Rice['low'] = fuzz.trimf(Rice.universe, [300, 350, 400])
Rice['avg'] = fuzz.trimf(Rice.universe, [400, 500, 600])
Rice['high'] = fuzz.trimf(Rice.universe, [700, 800, 900])
Rice['veryhigh'] = fuzz.trimf(Rice.universe, [900, 950, 1000])

power['verylow'] = fuzz.trimf(power.universe, [10, 15, 20])
power['low'] = fuzz.trimf(power.universe, [15, 20, 25])
power['avg'] = fuzz.trimf(power.universe, [20, 30, 40])
power['high'] = fuzz.trimf(power.universe, [30, 50, 70])
power['veryhigh'] = fuzz.trimf(power.universe, [50, 70, 100])


# #----------------------------------------------------------------------------------
# #-------------rule-------------------------

rule1  = ctrl.Rule(Rice['verylow'] & Time['veryfast'], power['high'])
rule2  = ctrl.Rule(Rice['verylow'] & Time['fast'], power['avg'])
rule3  = ctrl.Rule(Rice['verylow'] & Time['avg'], power['low'])
rule4  = ctrl.Rule(Rice['verylow'] & Time['slow'], power['verylow'])
rule5  = ctrl.Rule(Rice['verylow'] & Time['veryslow'], power['verylow'])

rule6  = ctrl.Rule(Rice['low'] & Time['veryfast'], power['high'])
rule7  = ctrl.Rule(Rice['low'] & Time['fast'], power['avg'])
rule8  = ctrl.Rule(Rice['low'] & Time['avg'], power['low'])
rule9  = ctrl.Rule(Rice['low'] & Time['slow'], power['low'])
rule10  = ctrl.Rule(Rice['low'] & Time['veryslow'], power['verylow'])

rule11  = ctrl.Rule(Rice['avg'] & Time['veryfast'], power['high'])
rule12  = ctrl.Rule(Rice['avg'] & Time['fast'], power['high'])
rule13  = ctrl.Rule(Rice['avg'] & Time['avg'], power['avg'])
rule14  = ctrl.Rule(Rice['avg'] & Time['slow'], power['avg'])
rule15  = ctrl.Rule(Rice['avg'] & Time['veryslow'], power['low'])

rule16  = ctrl.Rule(Rice['high'] & Time['veryfast'], power['veryhigh'])
rule17  = ctrl.Rule(Rice['high'] & Time['fast'], power['high'])
rule18  = ctrl.Rule(Rice['high'] & Time['avg'], power['high'])
rule19  = ctrl.Rule(Rice['high'] & Time['slow'], power['avg'])
rule20  = ctrl.Rule(Rice['high'] & Time['veryslow'], power['low'])

rule21  = ctrl.Rule(Rice['veryhigh'] & Time['veryfast'], power['veryhigh'])
rule22  = ctrl.Rule(Rice['veryhigh'] & Time['fast'], power['veryhigh'])
rule23  = ctrl.Rule(Rice['veryhigh'] & Time['avg'], power['high'])
rule24  = ctrl.Rule(Rice['veryhigh'] & Time['slow'], power['avg'])
rule25  = ctrl.Rule(Rice['veryhigh'] & Time['veryslow'], power['avg'])



# #------------------------------------------------------------------------------------
# #---------------training--------------------------------------------------

rule = [rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9, rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18, rule19, rule20, rule21, rule22, rule23, rule24, rule25  ]
tipping_ctrl = ctrl.ControlSystem(rule)
powering = ctrl.ControlSystemSimulation(tipping_ctrl)
powering.input['Time'] = 170
powering.input['Rice'] = 1000
powering.compute()
print(powering.output['power'])
power.view(sim = powering)
```
