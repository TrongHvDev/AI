```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sas
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Housing.csv')
data.hist()
plt.show()
```

```python
data.corr()
```

```python
sas.heatmap(data.corr(), annot = True)
```

Gia tri price phu thuoc lon nhat vao area

```python
data.plot(kind = 'scatter', x = 'area', y = 'price', figsize = (9, 9), color = 'black')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame(data['area'])
y = data['price']    
model.fit(X, y)
print('Intercept của mô hình bậc 2: ')
print(model.intercept_)
print('Coeffient của mô hình bậc 2: ')
print(model.coef_)
print('Độ chính xác của mô hình: ')
print(model.score(X, y))
data.plot(kind = 'scatter', x = 'area', y = 'price', figsize = (9, 9), color = 'black')
y_pred = model.predict(X)
plt.plot(data['area'], y_pred, color = 'red')
```

# bac 2

```python
model = linear_model.LinearRegression()
X_2 = pd.DataFrame([data['area'], data['area'] **2]).T    # a1x + a2x^2 + b, .T la nghich dao
model.fit(X_2, y = data['price'])
print('Intercept của mô hình bậc 2: ')
print(model.intercept_)
print('Coeffient của mô hình bậc 2: ')
print(model.coef_)
print('Độ chính xác của mô hình: ')
print(model.score(X_2, y))
range = np.arange(2000, 16000, 20)
predictor = pd.DataFrame([range, range **2]).T
y_pred = model.predict(X = predictor)
data.plot(kind = 'scatter', x = 'area', y = 'price', figsize = (9, 9), color = 'black')
plt.plot(range, y_pred, color = 'red')
```
